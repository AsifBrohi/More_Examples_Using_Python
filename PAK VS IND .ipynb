```python
import pandas as pd 
from matplotlib import pyplot as plt 
sales_10000 = pd.read_csv('10000_sales.Csv')
```

```python
sales_10000
```

```python
PAK = sales_10000[sales_10000.Country == 'Pakistan']
IND = sales_10000[sales_10000.Country == 'India']
```

```python
PAK
```

```python
plt.barh(PAK.Item_Type, PAK.Units_Sold, color = 'b')

plt.legend(['Pakistan'])
plt.xlabel('Units Sold')
plt.ylabel('Item Type')
plt.title('Pakistan = Item type vs Units Sold')
plt.show()
```

```python
plt.barh(IND.Item_Type, IND.Units_Sold, color = 'c')
plt.legend(['India'])
plt.xlabel('Units SOld')
plt.ylabel('Item Type')
plt.title('India = Item type vs Units sold')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_groups = 6
means_frank = (90, 55, 40, 65, 78, 32)
means_guido = (85, 62, 54, 20, 88, 45)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, means_frank, bar_width,
alpha=opacity,
color='b',
label='Frank')

rects2 = plt.bar(index + bar_width, means_guido, bar_width,
alpha=opacity,
color='g',
label='Guido')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D', 'E', 'F'))
plt.legend()

plt.tight_layout()
plt.show()
```
